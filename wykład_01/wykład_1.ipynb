## 1. Czym jest wstępne przetwarzanie danych (ang. data preprocessing, data wrangling)?

Proces wstępnego przetwarzania danych może być wieloetapowy i złożony, ale jego głównym celem jest przygotowanie danych do kolejnego etapu ich wykorzystania, np. do sporządzenia analizy, zestawienia czy przygotowania danych do stworzenia modelu machine learning.


**Wstępne przetwarzanie danych może polegać na:**
* poprawianie błędów w danych,
* uzupełnianie wartości brakujących (ang. missing values),
* tworzenie nowych cech z dostępnych danych,
* przekształcenia takie jak skalowanie, normalizacja, standaryzacja,
* enkodowanie danych nominalnych, porządkowych na dane numeryczne,
* redukcja wymiarów,
* oczyszczanie danych z szumu,
* przekształcenia danych tekstowych pod kątem zadań przetwarzania języka naturlanego (ang. NLP - Natural Language Processing) czyli tokenizacja, usuwanie słów stop, stemming, lematyzacja,
* w przypadku danych obrazowych również augmentacja, przycięcie, zmiana rozdzielczości itp.

I tu warto przytoczyć krótką, lecz bardzo prawdziwą maksymę.

<p style="font-size: 2em; font-style: italic; text-align: center; padding: 50px 0;">Garbage in -> Garbage out</p>

Oznacza to, że jakiej jakości dane "włożymy" do modelu, takie uzyskamy wyniki.

Wiadomo również, że ilość danych ma wpływ na model jednak wiele doświadczeń wskazuje, że 5% lepszych danych może dać lepszą poprawę modelu niż dodatkowe 500% danych.

Można więc pokusić się o parafrazę stwierdzenia **"Gospodarka, głupcze!"** ukutego przez Jamesa Carville’a dla Billa Clintona na **"Jakość danych, głupcze!"**.

**Źródła problemów z danymi:**
* **błędy pomiaru:**
  * błędne dane ze względu na uszkodzenie urządzenia (czujnika, kamery, itp.),
  * brakujące wartości ze względu na uszkodzenie urządzenia,
  * błąd algorytmu OCR, artefakty na obrazach,
* **błędy "ludzkie":**
  * literówki i pomyłki we wprowadzaniu danych (np. w formularzach),
  * błędne etykiety (np. w zbiorze obrazów z roślinami, zwierzętami, pojazdami, itp.),
* **błędy przetwarzania:**
  * błędna konwersja formatów,
  * błędy eksportu/importu (np. kodowanie znaków),
  * błędy agregacji danych,
  * niespójna jednostka dla cechy, np. kg i funty,
* **nieścisłość danych:**
  * obciążenie (ang. bias) - np. trening modelu na obrazach młodych ludzi, a predykcja wykonywana jest na obrazach osób starszych,
  * nieaktualne dane (nie odzwierciedlaja aktualnej rzeczywistości),
* **wartości odstające (ang. outlier)** czyli takie, których wartość znacznie odbiega od pewnej normy danych (wynikającej np. z właściwości rozkładu danych dla danej cechy).


| **Źródło błędu**    | **Przykład**            | **Wpływ**             |
| --------------- | -------------------- | --------------------- |
| pomiar          | uszkodzony czujnik   | błędne wartości       |
| człowiek        | literówki            | niespójność           |
| brak danych     | puste pola           | trudności w uczeniu   |
| processing      | zła konwersja        | błędne dane           |
| sampling        | bias                 | model stronniczy      |
| outliery        | ekstremalne wartości | destabilizacja modelu |
| integracja      | różne jednostki      | błędy                 |
| stare dane      | nieaktualne trendy   | złe predykcje         |
| noise           | zakłócenia           | gorsza dokładność     |
| błędne etykiety | zła klasa            | model uczy się źle    |


## 2. Typy danych

Typy danych to sposób klasyfikacji danych według ich charakteru matematycznego i znaczenia. To bardzo ważne w analizie danych i machine learning, bo różne typy wymagają różnych metod przetwarzania.

Podstawowy podział wygląda tak:

```
Dane
├── Numeryczne (ilościowe, ang. quantitative)
│   ├── dyskretne
│   └── ciągłe
└── Kategoryczne (jakościowe, ang. qualitative)
    ├── nominalne
    └── porządkowe (ang. ordinal data)
```

### 2.1. Dane numeryczne (ilościowe)
To dane wyrażone liczbami, na których można wykonywać działania matematyczne. Warto wspomnieć również, że w ostatecznej formie dane, które staną się zbiorem danych uczących dla algorytmów machine learning zawsze przyjmą formę numeryczną, bez względu na to czy jest to zdanie, rekord z bazy danych, obraz lub dźwięk.

#### 2.1.1. Dane dyskretne

Przyjmują konkretne, oddzielne wartości (najczęściej całkowite).

**Przykłady:**

* liczba dzieci: 0, 1, 2, 3
* liczba klientów
* liczba błędów w kodzie
* liczba pikseli

#### 2.1.2. Dane ciągłe

Mogą przyjmować dowolną wartość z zakresu (także ułamki).

**Przykłady:**

* wzrost: 175.3 cm
* waga: 72.8 kg
* temperatura: 36.6 °C
* czas reakcji: 0.235 s

Te dane pochodzą zwykle z pomiarów.

### 2.2. Dane kategoryczne (jakościowe)

Reprezentują kategorie, nie liczby w sensie matematycznym.

#### 2.2.1. Dane nominalne

Kategorie bez kolejności.

Nie ma sensu mówić, że jedna jest „większa” od drugiej.

**Przykłady:**

* kolor: czerwony, zielony, niebieski
* miasto: Warszawa, Kraków, Gdańsk
* płeć: mężczyzna, kobieta
* typ zwierzęcia: kot, pies, ptak

#### 2.2.2. Dane porządkowe

Kategorie mają kolejność, ale różnice nie są matematycznie mierzalne.

**Przykłady:**

* **poziom wykształcenia:**
  * podstawowe
  * średnie
  * wyższe
* **wzrost (np. portret pamięciowy):**
  * niski
  * wysoki
  * bardzo wysoki
* **oceny w szkole:**
  * 2, 3, 4, 5

Wiemy, że:

dobry > średni > słaby

ale nie wiemy na tej podstawie "o ile".



#### 2.2.3 Dane tekstowe

To specjalny typ danych kategorycznych, który reprezentowany jest przez zdania, z których składają się dokumenty.

Takie dane muszą być wcześniej zamienione na liczby, które ten tekst reprezentują. Proces jest wieloetapowy i polega na wykorzystaniu m.in. tokenizacji czy stemmingu.

#### 2.2.4 Dane binarne

Dane binarne czyli dane z dwoma wartościami, są szczególnym przypadkiem danych kategorycznych.

**Przykłady:**

* tak / nie
* 0 / 1
* prawda / fałsz

### 2.3 Szeregi czasowe

Nie jest to oddzielna kategoria w odniesieniu do danych ilościowych czy jakościowych, jednak często traktowana jako podkategoria. 

Te dane charakteryzują się połączeniem wartości konkretnej obserwacji z czasem, wg. którego są one uszeregowane.

**Przykłady:**

* ceny akcji
* temperatura w czasie
* sygnał EEG

Możemy jeszcze podzielić dane na typy według ich struktury:

* **Dane strukturalne:** Uporządkowane w wierszach i kolumnach (relacyjne bazy danych SQL).
* **Dane półstrukturalne:** Mają etykiety, ale nie sztywną strukturę (pliki JSON, XML).
* **Dane nieustrukturyzowane:** Pliki binarne, obrazy, wideo, swobodny tekst (np. maile, komentarze).

W kontekście narzędzi opartych o język Python oraz R wykorzystywany jest termin ramka danych (nag. data frame). Jest to tabela, która zawiera rekordy (wiersze) oraz cechy (kolumny) opisujące te dane. Poniżej kilka terminów, które warto poznać, gdyż są w literaturze używane zamiennie.

| Nazwa | Opis | Synonimy |
| --- | --- | --- |
***Ramka danych*** | Stabelaryzowany zbiór danych | 
***Cecha*** | Kolumna w ramce danych | atrybut, wejście (ang. input), predyktor, zmienna
***Wynik*** | To wartość, która w jakiś sposób podsumowuje lub kategoryzuje dany rekord, np. chory/zdrowy, kot/pies/człowiek, 123234 zł/345678 zł (regresja). | zmienna zależna, odpowiedź, cel (ang. target), wyjście (ang. output)

####  3.2.1 Miary położenia

##### Średnia

Suma wszystkich wartości podzielona przez liczbę tych wartości.

**Wzór:**

$$\huge \bar{x} = \frac{\sum_{i=1}^{n} x_i}{n}$$

**Cechy/zastosowanie:**

* wrażliwa na wartości skrajne oraz skośność danych,
* taka miara jest często używana do oceny uczniów w szkole, gdzie poszczególnym ocenom przypisywane są różne wagi (np. praca klasowa, praca domowa),
* można również wykorzystywać wagi przy wyliczaniu metryk dla niedoreprezentowanych grup w danych.


##### Średnia ważona

Suma wartości wymnożonych przez wagi podzielona przez sumę tych wag.

**Wzór:**

$$\huge \bar{x} = \frac{\sum_{i=1}^{n} w_i x_i}{\sum_{i=1}^{n} w_i}$$


##### Średnia ucinana

Średnia powstała po odrzuceniu stałej liczby wartości skrajnych ze zbioru.

**Wzór:**

$$\huge \bar{x} = \frac{\sum_{i=p+1}^{n-p} x_{(i)}}{n - 2p}$$

Oznacza to, że po posortowaniu zbioru odrzucamy ***p*** elementów ze skrajów posortowanego zbioru (czyli ***p*** najmniejszych oraz ***p*** największych).

**Cechy/zastosowanie:**
* mniejszy wpływ skrajnych danych na wartość średniej,
* w niektórych zawodach sportowych dwie skrajne noty są odrzucane i z wartości pozostałych wyliczana jest średnia nota, aby zredukować efekt ewentualnej stronniczości sędziów.


##### Mediana

Wartość, dla której połowa danych jest od niej mniejsza, a druga połowa jest od niej większa. nazywana również wartością środkową. Jest to 50 percentyl.

**Wzór:**

Wyliczenie mediany zależy od tego czy zbiór zawiera parzystą czy nieparzystą liczbę elementów.

Dla nieparzystej liczby elementów posortowanego zbioru danych medianę wyznaczamy wzorem:

$$\huge mediana = \left (\frac{n + 1}{{2}} \right)^{ta}\ wartość$$

Dla parzystej liczby elementów posortowanego zbioru danych medianę wyznaczamy wzorem:


$$\huge mediana = \frac{(\frac{n}{2})^{ta}\ wartość + (\frac{n + 1}{2})^{ta}\ wartość}{{2}}$$

**Cechy/zastosowanie:**
* nie jest wrażliwa na wartości odstające, nazywana odpornym (ang. robust) estymatorem położenia,
* często stosowana do wyliczenia średniego (właściwie najczęstszego) wynagrodzenia w danej grupie.

####  3.2.2 Miary rozproszenia

Miary rozproszenia pozwalają nam zbadać jak daleko dane "leżą" od siebie w całym zbiorze, ale również danych podgrupach. Omówimy wybrane miary poniżej.

##### Odchylenie

Jest to różnica między wartością obserwowaną a szacowanym położeniem (wartością oczekiwaną).

Inne terminy, którą są używane do jej określenia to: **błąd, rezyduum, reszta.**

**Średnie odchylenie bezwzględne (MAD - Mean Absolute Deviation)**

**Wzór:**

$$\huge MAD = \frac{\sum^n_{i=1}|x_i - \bar{x}|}{{n}}$$

Dzięki temu różnice względne wartości oczekiwanej względem wartości średniej nie znoszą się wzajemnie dla dodatnich i ujemnych wartości różnic.

**Synonimy:**
* norma L1
* norma Manhattan

##### Zakres, rozstęp

Jest to różnica między wartością minimalną oraz maksymalną danych.
Tu raczej wzór potrzebny nie jest.

##### Wariancja

jest to suma kwadratów odchylenia od średniej podzielonych przez *n - 1* lub *n*, gdzie *n* to liczba wartości w danych. Różnica w mianowniku powiązana jest z zagadnieniem stopni swobody.

**Wzór dla populacji:**

$$\huge  wariancja = \sigma^2 = \frac{\sum^n_{i=1}(x_i - \mu)^2}{{n}}$$

**Wzór dla próby:**

$$\huge  wariancja = \sigma^2 = \frac{\sum^n_{i=1}(x_i - \bar{x})^2}{{n - 1}}$$

W obu przypadkach liczymy sumę różnicy między wartością oczekiwaną a średnią (czy to oznaczone przez $\mu$ czy poprzez $\bar{x}$). Wzór dla populacji nazywany jest też **estymatorem obciążonym (ang. biased estimator)**, a dla próby **estymatorem nieobciążonym** poprzez zmianę w podejściu do wyznaczenia mianownika związanego z zagadnieniem stopni swobody.

**Stopnień swobody** to liczba niezależnych ruchów / obserwacji/ wyników jakiej operacji. Dla przykładu dla wyboru jednej z 4 kart, mamy dla pierwszego ruchu 4 możliwości, później 3, 2, a dla ostatniej wyboru już nie ma. Liczba stopni swobody wynosi więc 4 - 1 = 3. 

Ale skąd takie podejście w statystyce?

Tutaj chodzi o liczbę parametrów równania, które są niezależne. W przypadku wariancji czy odchylenia standardowego mamy jedną zmienną zależną, średnią, która jest wcześniej wyliczona. Jeżeli chcemy, aby ten współczynnik nie był obciążony (błędem), to musimy od liczby obserwacji odjąć liczbę tych zależnych zmiennych.

Dla wybranych wzpółczynników liczbę stopni swobody możemy wyliczyć zgodnie z zestawieniem: https://www.naukowiec.org/tablice/statystyka/stopnie-swobody-dla-testow-statystycznych_378.html

Intuicyjnie rzecz ujmując, jeżeli nie znamy faktycznej wartości średniej dla populacji to licząc wskaźnik (tu odchylenia standardowego) szacujemy średnią, która jest potrzebna do wyznaczenia odchylenia, na podstawie danych z próby, a nie populacji. Te wartości mogą się różnić! Im mniejsza, i mniej reprezentatywna próba danych względem populacji, tym te wyliczenia mogą bardziej się różnić.

Czasami jednak nie jesteśmy w stanie zebrać danych z całej populacji.

Możesz przeczytać dodatkowo:
* https://www.statystyczny.pl/odchylenie-standardowe-z-proby/

Wariancja jest kwadratem odchylenia standardowego ($\sigma^2$) co utrudnia nieco jej interpretację w stosunku do danych, dla których jest wyliczona. Należy o tym pamiętać. Z pomocą przychodzi odchylenie standardowe, które wyrażone jest w tych samych jednostkach, w których dane, na podstawie, których zostało wyliczone.

##### Odchylenie standardowe

**Wzór:**

$$\huge odchylenie \ standardowe = \sigma = \sqrt{wariancja}$$ 

**Percentyl, Centyl, Kwantyl, Decyl itd.**

W zbiorze danych P-ty percentyl, np. 25-ty percentyl oznacza, że co najmniej P procent wartości jest od niego mniejszych lub równych oraz co najmniej (100-P) procent wartości jest równch lub większych od niego. Podobnie jak przy wyliczaniu mediany wartości muszą być uporządkowane (posortowane).

Percentyle są wartościami określającymi procent wartości po lewej jego stronie (dane posortowane rosnąco) w przedziale <0,100>. Kwantyle są bardzo podobnym pojęciem, ale tu wartość podajemy z przedziału <0,1>. 50-ty percentyl jest równy medianie i kwantylowi 0.5.

Można dla lepszej intuicji przyjąć taki podział i nazewnictwo:

**Kwantyl** rzędu $q$ (gdzie $0 \lt q \lt 1$) to taka wartość $x_q$, poniżej której znajduje się $100q\%$ obserwacji.

**Zastosowanie:**  
Służą do analizy rozkładu danych, porównywania wyników i identyfikacji wartości odstających.

**Rodzaje kwantyli:**
* **Mediana** (kwantyl rzędu 0,5): Dzieli dane na pół (50% danych jest mniejszych lub równych).
* **Kwartyle** (podział na 4 części): Dolny ($Q_1 = 25\%$), Mediana ($Q_2 = 50\%$), Górny ($Q_3 = 75\%$).
* **Decyle** (podział na 10 części): $D_1$ (10%), $D_2$ (20%), ..., $D_9$ (90%).
* **Percentyle/Centyle** (podział na 100 części): $P_1$ (1%), $P_2$ (2%), ..., $P_{99}$ (99%).

**Interpretacja:**

Jeżeli Twoja waga znajduje się w $P_{30}$ to oznacza, że 30% osób w danej populacji waży mniej lub tyle samo co Ty przy danych parametrach (wiek, wzrost).
Możemy zatem wyliczać, w którym miejscu (percentyl, kwartyl, decyl, itd.) znajduje się dana wartość, ale również wyliczać wartość, która odpowiada podanemu położeniu (mediana, 75-ty percentyl, itd.). Jest to również ściśle powiązane z rozkładami danych, które zostaną przedstawione w kolejnych wykładach.

**Przedział międzykwartylowy (IQR):** 

Zakres między 25. a 75. percentylem, obejmujący środkowe 50% danych.

**Wzór (percentyl):**

$$\huge  100 * \frac{j}{n} \le P \le 100 * \frac{j + 1}{n}$$
$$\huge  percentyl(P) = (1 - w)x_{(j)} + wx_{(j+1)}$$


Możesz też zobaczyć wyjaśnienie działania (również przetestować) oraz obliczania percentyla na stronie: https://www.calculatorsoup.com/calculators/statistics/percentile-calculator.php

#### Percentyle/Kwantyle/Decyle

````python
# ten moduł posiada wbudowaną funkcję quantiles, która dzieli dane na przedziały
# domyślnie są 4 (czyli kwartyle 0.25, 0.5, 0.75). Dane są rodzielane tak, że
# prawdopodobieńtwo dla każdego kwantyla jest równe (w miarę możliwości jego wyliczenia)
# Wynik, który otrzymaliśmy jest granicą przedziałów (kwartyli), oznacza to, że 25% wartości jest mniejszych lub równych -11.25,
# 25 wartości leży w przedziale [-11.25, -5.5] itd.
# Jeżeli stworzymy 10 przedziałów otrzymamy decyle, 100 - percentyle
kwartyle = statistics.quantiles(data)
kwartyle
````

````python
# wylosujmy ze zwracaniem 100 wartości z danych i sprawdźmy ile z nich znalazło się w pierwszym kwartylu
samples = random.choices(data, k=100)
````

````python
# procent wylosowanych liczb, które znajdują się w Q1
len(list(filter(lambda x: x <= kwartyle[0], samples))) / len(samples)
````

````python
# bardzo blisko 25%, tu również zadziała teoria wielkich liczb, czyli im bardziej będziemy zwiąkszać wielkość zbioru próbek
# tym bardziej rozkład będzie zbliżał do rozkładu populacji
samples = random.choices(data, k=1_000_000)
len(list(filter(lambda x: x <= kwartyle[0], samples))) / len(samples)
````

Czy wartość kwartyle Q2 z powyższego przykładu nie pojawiła się już we wcześniejszych wynikach dla podstawowych statystyk?
